```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    if(x< -10 or x > 10):
        return -np.inf
    else:
        return 100 - x**2
```

```python
def DeltaX(amplitud):
    return amplitud*np.random.rand() - (amplitud/2.0)
```

```python
N = 100000
```

```python
walk = []
walk.append(np.random.uniform(-10,10))
count = 0
ancho = 0.0005
while count < N:
    x_old = walk[-1]
    dx = DeltaX(ancho)
    x_new = x_old + dx
    alpha = func(x_new)/func(x_old)
    if (alpha >= 1.0):
        walk.append(x_new)
        count += 1
    else:
        beta = np.random.random()
        if(beta <= alpha):
            walk.append(x_new)
            count += 1
        else:
            walk.append(x_old)
            count += 1
            
```

```python
plt.plot(walk)
```

```python
plt.hist(walk, bins=np.linspace(-12,12,51))
```
